```julia
function my_factorial(n)
    out = 1
    for i in 2:n
        out *= i
    end
    out
end

my_factorial(big(10))
# @time my_factorial(big(20))
```

```julia
T = 100
x = zeros((3,T))
epsilon = randn((3,T))
alpha = [0; 0.5; 0.9]
for i in 2:T
    x[:,i] = alpha .* x[:,i-1] .+ epsilon[:,i]
end

plot(x', labels=["0" "0.5" "0.9"], lw=3)
```
